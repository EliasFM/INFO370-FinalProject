```python
# split year-month into a month column and a year column
df['year'] = df.month.str.slice(0,4)
df.month = df.month.str.slice(5,)
```

```python
df.head()
```

```python
# remove subcategories of how different guntypes are sold
df= df[['month','state','permit','handgun','long_gun','other','multiple','admin','totals','year']]
```

```python
# convert month and year into ints and remove all years before 2013 (where we have mass shooting data)
df.month = df.month.apply(lambda x:int(x))
df.year =df.year.apply(lambda x:int(x))
df = df[df.year>2012]
```

```python
# we can see that we have no null values now
df.isnull().sum()
```

```python
df.head()
```

```python
#add United States
national = df.groupby(['year','month'])[['permit','handgun','long_gun','other','multiple','admin','totals']].sum().reset_index()
national['state'] = 'United States'
df = df.append(national)
```
